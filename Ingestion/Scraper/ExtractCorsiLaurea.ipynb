```python
import requests
import pandas as pd
import time

# URL base con segnaposto per la pagina
base_url = (
    "https://universitaly-backend.cineca.it/api/offerta-formativa/cerca-corsi?"
    "searchType=u&page={page}&tipoLaurea&tipoClasse=0&durata&lingua&tipoAccesso&"
    "modalitaErogazione&searchText&area&order=ASC&provincia&provinciaSigla"
)

# TODO recupeara numero max di pagine dal json
total_pages = 580

# Lista per raccogliere tutti i corsi "raw"
all_corsi = []

for page in range(1, total_pages + 1):
    print(f"Scarico pagina {page}/{total_pages}")
    
    url = base_url.format(page=page)
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()
        corsi = data.get("corsi", [])

        print(f"  ➤ Trovati {len(corsi)} corsi")

        # Aggiunge ogni corso così com'è, senza flat
        all_corsi.extend(corsi)

        time.sleep(0.3)

    except Exception as e:
        print(f"⚠️ Errore a pagina {page}: {e}")
        continue

# Converti tutti i corsi in DataFrame
df = pd.json_normalize(all_corsi)

# Salva su CSV
df.to_csv("universitaly_corsi_completi_raw.csv", index=False)

print(f"\n Totale corsi salvati: {len(df)}")
```
