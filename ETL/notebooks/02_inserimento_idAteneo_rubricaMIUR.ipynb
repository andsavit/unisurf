```python
import pandas as pd

# Carico i due dataframe
# file1 = pd.read_csv("file1.csv")
# file2 = pd.read_csv("file2.csv")

# facciamo il merge usando NomeEsteso come chiave
merged = file2.merge(
    file1[["NomeEsteso", "ror", "id_oa"]],
    on="NomeEsteso",
    how="left",
    suffixes=("", "_from_file1")
)

# se file2 ha già colonne ror / id_oa vuote da riempire, possiamo copiarle
merged["ror"] = merged["ror"].combine_first(merged["ror_from_file1"])
merged["id_oa"] = merged["id_oa"].combine_first(merged["id_oa_from_file1"])

# eliminiamo le colonne extra
merged = merged.drop(columns=["ror_from_file1", "id_oa_from_file1"])

# salva su csv se vuoi
merged.to_csv("file2_enriched.csv", index=False, encoding="utf-8")
```
