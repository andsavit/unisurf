```python
import pandas as pd
```

```python
atenei_miur=pd.read_csv("data/raw_data/miur/ListaAteneiMUR.csv", sep=',')
ponte = pd.read_csv("ETL/TabellePonte/ponte_OA_MIUR_test", sep=';')
docenti_statali=pd.read_csv("data/raw_data/miur/ListaDOCENTI_AteneiStatali.csv", sep=';')
ponteMatch = pd.read_csv("ETL/Trasformazione/PonteMatchAtenei.csv")
```

```python
joined = pd.merge(atenei_miur, ponte, on = 'NomeEsteso', how='left') 
joined
```

```python
joined2=pd.merge(joined, ponteMatch, left_on='NomeOperativo', right_on = 'NomeOperativo', how='left')
joined2
```

```python
len(joined2) - sum(joined2['id_oa'].isna())
print()
```

```python
joined2.columns
```

```python
enriched = joined2[['ror', 'id_oa', 'NomeEsteso', 'NomeOperativo', 'status', 'Tipologia_ateneo_descrizione', 'Statale/NonStataleLibera', 'Indirizzo', 'COMUNE', 'PROVINCIA', \
                   'REGIONE', 'Zona Geografica', 'URL', 'Classe dimensione', 'NomeRubrica']] 
enriched
```

```python
enriched.columns
enriched = enriched.rename(columns={"Tipologia_ateneo_descrizione":"Tipologia", "Statale/NonStataleLibera": "Statale", "Zona Geografica": "ZonaGeografica", "Classe dimensione":"ClasseDimensione"})
```

```python
enriched.columns
```

```python
enriched.to_csv("data/tabelle/AteneiEnriched.csv", index=False)
```
