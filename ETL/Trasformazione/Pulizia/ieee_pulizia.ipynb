```python
from pymongo import MongoClient
```

```python
client = MongoClient("mongodb://localhost:27017/")

# Seleziona database e collezione
db = client["unisurf"]
collection = db["ieee_puliti"]
```

```python
pipeline = [
    {"$group": {"_id": "$doi", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}},
    {"$limit": 10}  # opzionale, rimuovi se vuoi tutti
]
```

```python
results = collection.aggregate(pipeline)

for doc in results:
    print(doc)
```

```python
#Trova le distinte università e ne conta l'occorrenza
pipeline = [
    {"$unwind": "$authors.authors"},
    {"$group": {"_id": "$authors.authors.affiliation", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}}
]

results = collection.aggregate(pipeline)

for doc in results:
    print(doc)
```
