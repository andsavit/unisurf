```python
client = MongoClient("mongodb://localhost:27017/")
db = client['unisurf']
collection = db['ieee_puliti']
```

```python
pipeline = [
    {"$match": {"doi": {"$exists": True, "$ne": None}}},   # tieni solo documenti con campo doi
    {"$group": {"_id": None, "all_doi": {"$addToSet": "$doi"}}}  # raccogli tutti i doi in una lista (set elimina i duplicati)
]

result = list(collection.aggregate(pipeline))

# Estrai la lista
doi_list = result[0]["all_doi"] if result else []

print(len(doi_list))
```

# Controllo Formato autori e istituzioni

```python
pipeline = [
    {"$unwind": "$authors"},             # se 'authors' è un oggetto
    {"$unwind": "$authors.authors"},     # scendi nell'array 'authors'
    {"$group": {
        "_id": None,
        "distinct_authors": {"$addToSet": "$authors.authors.full_name"}
    }}
]

result = list(collection.aggregate(pipeline))

# Estrai lista
author_names = result[0]["distinct_authors"] if result else []
print(author_names)
```

```python
for el in author_names:
    print(el)
```

```python
pipeline = [
    {"$unwind": "$authors"},
    {"$unwind": "$authors.authors"},
    {"$group": {
        "_id": "$authors.authors.affiliation",
        "count": {"$sum": 1}
    }},
    {"$sort": {"_id": 1}}   # ordine alfabetico per affiliazione (_id)
]

result = list(collection.aggregate(pipeline))

# Converto in lista di tuple (affiliazione, occorrenze)
affiliations = [(doc["_id"], doc["count"]) for doc in result if doc["_id"]]

for aff, c in affiliations:
    print(f"{aff}: {c}")
```
